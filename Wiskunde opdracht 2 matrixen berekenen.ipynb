```python
import numpy as np

# Parameters
l = 0.6         
g = 9.81         
rho_wood = 700  

# Massa berekenen
breedte = 0.02
hoogte = l
diepte = 0.02


V = breedte * hoogte * diepte
m_p = rho_wood * V # massa pendulum
ml2 = m_p * l**2 # moeten we veel gebruiken in de berekeningen# Afmetingen pendulum

Icm = (1/12) * m_p * l**2
I = (1/3) * m_p * l**2
# Matrix M
M = np.array([[I+(3/2 * ml2), 1/2 * ml2],
              [Icm+(3/4 * ml2), Icm+(1/4 * ml2)]])

# Inverse matrix M
M_inv = np.linalg.inv(M)

# Matrix G
G = np.array([[3/2 * m_p * g * l, 0],
              [1/2 * m_p * g * l, 1/2 * m_p * g * l]])

# Matrix H
H = np.array([[1],
                [0]])

# Bereken M^-1 * G en M^-1 * H
MG = M_inv @ G
MB = M_inv @ H

# theta, theta', phi, phi' in matrix zetten
A = np.array([[0, 1, 0, 0],
              [MG[0,0], 0, MG[0,1], 0],
              [0, 0, 0, 1],
              [MG[1,0], 0, MG[1,1], 0]])

B = np.array([[0],
              [MB[0,0]],
              [0],
              [MB[1,0]]])

#matrixen geven
print("A-matrix:",A)
print("B-matrix:",B)
print("De eigenwaardes van A zijn", np.linalg.eigvals(A))
```
