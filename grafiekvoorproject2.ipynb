```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
gegevens = {
    'Imput-cost':[0.006746499885963993, 0.005439650645699048, 0.005392660549708741, 0.005699785050928485, 0.006146914852113442, 0.006664454094721414, 0.0072241022443392386],
    'Time':[3.546, 3.010, 2.538, 1.722, 1.682, 1.502, 1.306]
}
 
data = pd.DataFrame(gegevens)

colors= ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink']
names= ['[-3,-4,-5,-6]', 
        '[-4,-5,-6,-7]',
        '[-5,-6,-7,-8]',
        '[-6,-7,-8,-9]',
        '[-7,-8,-9,-10]',
        '[-8,-9,-10-,11]',
        '[-9,-10,-11,-12]']

y_data = np.array(data['Imput-cost'])
x_data = np.array(data['Time'])


plt.figure(figsize=(10, 6))



for i in range(len(x_data)):
    plt.scatter(x_data[i], y_data[i], color=colors[i])
    plt.annotate(names[i], xy=(x_data[i], y_data[i]), xytext=(4, 6), textcoords="offset points", ha='center', fontsize=9)

plt.title('Imput-cost against Time')
plt.xlabel('Time (s)')
plt.ylabel('Imput-cost')
plt.legend()
```
